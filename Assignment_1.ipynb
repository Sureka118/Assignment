```python
#to perform univariate analysis 

#price frequency

import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt

# Load house price data from CSV file
house_prices_df = pd.read_csv('House Price India.csv')

# Extract house prices as a NumPy array
house_prices = house_prices_df['Price'].values

# Create histogram
plt.hist(house_prices, bins=10, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')

# Show the plot
plt.show()
```

```python
#univariate analysis
# price range in boxplot

import matplotlib.pyplot as plt
import pandas as pd

# Read in the data
data = pd.read_csv('House Price India.csv')

# Extract the column of interest
column_of_interest = data['Price']

# Create a box plot
fig, ax = plt.subplots()
ax.boxplot(column_of_interest)
ax.set_title('Box plot of {}'.format('Price'))
ax.set_ylabel('Values')
plt.show()
```

```python
#uni-variate analysis
# histogram plot for the frequency of house prices
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('House Price India.csv')

# Choose the price column
prices = df['Price']

# Set the number of bins
num_bins = 20

# Plot the histogram
plt.hist(prices, bins=num_bins)

# Add labels and title
plt.xlabel('House Price')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')

# Show the plot
plt.show()

# Identify the most frequent price
bin_counts, bin_edges, _ = plt.hist(prices, bins=num_bins)
most_frequent_price = bin_edges[bin_counts.argmax()]
print(f"The most frequent house price is {most_frequent_price}")
```

```python
#bi-variate analysis
# price with number of floors

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('House Price India.csv')

# Extract two columns to compare
col1 = df['number of floors']
col2 = df['Price']

# Create scatter plot
plt.scatter(col1, col2)

# Add labels and title
plt.xlabel('number of floors')
plt.ylabel('price')
plt.title('Comparison of number of floors and price')

# Show the plot
plt.show()
```

```python
#multivariate analysis
#comparing number of bedroom  and  number of bathroom with price

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('House Price India.csv')

# Extract columns to compare
col1 = df['number of bedrooms']
col2 = df['number of bathrooms']
col3 = df['Price']

# Create figure with two y-axes
fig, ax1 = plt.subplots()

# Create scatter plot for col1 and col3
ax1.scatter(col3, col1, color='red')
ax1.set_xlabel('bedroom')
ax1.set_ylabel('Price', color='red')

# Create second y-axis for col2
ax2 = ax1.twinx()
ax2.scatter(col3, col2, color='blue')
ax2.set_ylabel('bathroom', color='blue')

# Add title
plt.title('Comparison of bedroom, bathroom, and price')

# Show the plot
plt.show()
```

```python
import pandas as pd

# load the dataset
data = pd.read_csv('House Price India.csv')

# select a column of interest
column_name = 'Price'

# calculate the descriptive statistics
mean = data['Price'].mean()
median = data['Price'].median()
std = data['Price'].std()
min_value = data['Price'].min()
max_value = data['Price'].max()
q1 = data['Price'].quantile(0.25)
q3 = data['Price'].quantile(0.75)

# print the results
print('Mean:', mean)
print('Median:', median)
print('Standard deviation:', std)
print('Minimum value:', min_value)
print('Maximum value:', max_value)
print('25th percentile:', q1)
print('75th percentile:', q3)

```
